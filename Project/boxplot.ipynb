```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"F:\INTERNSHIP\PYTHON\boxplot_data.xlsx"  
df = pd.read_excel(file_path)

print(df.head())

```

# Boxplot

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='churned', y='Task Completion Rates', data=df, palette="Set2")
plt.title("Boxplot of Task Completion Rates by Churned Employees")
plt.xlabel("Churned")
plt.ylabel("Task Completion Rates")
plt.show()
```

```python
numeric_columns = ['Days Present', 'Task Completion Rates', 'Interpersonal Skills Rating', 
                   'Decision-Making Skills Rating', 'Sales Revenue Generated']

fig, axes = plt.subplots(len(numeric_columns), 1, figsize=(8, 12))

for index, column in enumerate(numeric_columns):
    sns.kdeplot(df[column], ax=axes[index])
    axes[index].set_title(f"KDE Plot for {column}")

plt.tight_layout()
plt.show()
```
